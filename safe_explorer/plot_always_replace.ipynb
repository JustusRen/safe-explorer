```python
def average_dfs(dfs):
    # Concatenate all dataframes along the index
    df_concat = pd.concat(dfs)

    # Group by index and take the mean
    df_mean = df_concat.groupby(df_concat.index).mean()

    return df_mean
```

```python
average_losses_gs = average_dfs(losses_dfs_gs)
average_train_data_gs = average_dfs(train_data_dfs_gs)
average_eval_data_gs = average_dfs(eval_data_dfs_gs)
average_time_gs = average_dfs(time_dfs_gs)
```
